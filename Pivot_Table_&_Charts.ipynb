 3. Apply Conditional Formatting to highlight sales greater than 400.
    - First select the range of the data from D2 TO D21 from the sales column, after selecting the range of data go to "Home" tab and click on "Conditional Formatting" from style group, then click "Highlight Cells Rules", then select "Greater Than" from the submenu, then the new dialog box appears of conditional formatting enter 400 in the value box then select format with desired color then click "Ok" after clicking the ok all the sales value which is greater than 400 will get highlighted.





 4. Explain the difference between Sorting and Filtering with an example from this dataset.
    - Sorting and Filtering are both methods used to manage and arrange the data in excel but with different purposes, "Sorting" is used to rearrange or reorder all the rows in dataset according to the values in one or more columns for example, sorting the data by "Sales" would rearrange every row from the lowest sales value to the highest sales value, or highest sales value to lowest sales value, Sorting example( by sales)~ select all the data from the table (cell A1 to E21), then go to the "Data" tab then click sort in the sort and filter group, then sort dailog box appears then in sort by dropdown menu select "Sales", then in sort on dropdown menu select "Cell Values", then in Order dropdown menu select "Largest to Smallest" then click ok, "Filtering" is a way to sort information so that only see what want to see, filtering is used to hiding unwanted information, showing only proper or suitable data, Filtering example( by Mangoes)~ select the header row or cells A1 to E1 then go to the "Data' tab and click on Filter from sort and filter group, then dropdown arrows will appears in the header cells, then click on the filter arrow in the "Product" column header, in the filter menu uncheck the "select all" and check the mangoes option box then click ok, after clicking ok the rows where the product is mangoes will be visible, and all other rows will be hidden.
